### Part 2: Working with Lists (15 mins)

**Scenario:** You have a list of earthquake magnitudes recorded by a local seismometer over the last week.

**Data:** `[2.4, 3.1, 1.8, 4.5, 2.9, 0.5, 3.3]`

**Task:**
1. Create a list named `magnitudes` containing the data above.
2. Print the **first** magnitude in the list.
3. Print the **last** magnitude in the list (try using negative indexing).
4. Append a new reading of `5.1` to the list.
5. Print the total number of recordings (length of the list).

```python
# Part 2: Your Code Here

magnitudes = [2.4, 3.1, 1.8, 4.5, 2.9, 0.5, 3.3]

# Print first element

print('First:',magnitudes[0])

# Print last element

print('Last:',magnitudes[-1])

# Append 5.1

magnitudes.append(5.1)

# Print length of list

print('Length:',len(magnitudes))
```

### Part 3: Loops and Math (20 mins)

**Scenario:** You need to calculate the average magnitude of the earthquakes from Part 2. While Python has libraries for this, it is important to know how to do it manually first.

**Task:**
1. Initialize a variable `total_magnitude` to `0`.
2. Write a `for` loop that iterates through your `magnitudes` list.
3. Inside the loop, add each magnitude to `total_magnitude`.
4. After the loop, calculate the mean (average) by dividing the total by the number of earthquakes.
5. Print the average.

```python
# Part 3: Your Code Here

total_magnitude = 0
mlen = len(magnitudes)

# Write your for loop here:

for mag in range(0,mlen):
  total_magnitude = total_magnitude + magnitudes[mag]

# Calculate mean

avg_magnitude = (total_magnitude) / (mlen)

# Print result

print('Average:',avg_magnitude)
```

### Part 4: Conditional Logic (15 mins)

**Scenario:** You need to classify the earthquakes into categories to alert the public.
* **Magnitude < 3.0:** "Micro" (Not usually felt)
* **Magnitude >= 3.0:** "Minor" (Often felt)

**Task:**
1. Write a loop that goes through the `magnitudes` list.
2. Inside the loop, use an `if/else` statement to check the magnitude.
3. Print the magnitude and its classification (e.g., *"Magnitude 4.5 is Minor"*).

```python
# Part 4: Your Code Here

for mag in magnitudes:
    if mag < 3.0:
        print('Magnitude', mag, 'is Micro')
    else:
        print('Magnitude', mag, 'is Minor')

# I wanted to group them by category, so that's what I did below
# I also included a gap and title to visually seperate them

print('\nGrouped:')
cat_micro = []
cat_minor = []

for mag in magnitudes:
    if mag < 3.0:
        cat_micro.append(mag)
    else:
        cat_minor.append(mag)

print('Micro:', cat_micro)
print('Minor:', cat_minor)
```

### Part 5: Putting it all together




**Scenario:** Geothermal gradient calculation.
The temperature at the surface is 15°C. The temperature increases by 25°C for every kilometer of depth (Geothermal Gradient).

**Task:**
1. Create a list of depths in km: `[0, 1, 2, 3, 4, 5]`.
2. Use a loop to calculate the temperature at each depth.
3. Store these temperatures in a new list called `temperatures`.
4. Print the final list.

```python
# Part 5: Your Code Here

depths = [0, 1, 2, 3, 4, 5]
temperatures = []

surface_temp = 15
gradient = 25

# Loop through depths, calculate temp, and append to temperatures list

for depth in depths:
  temperatures_at_depth = surface_temp + (depth * gradient)
  temperatures.append(temperatures_at_depth)

print('Temperatures:',temperatures)
```
